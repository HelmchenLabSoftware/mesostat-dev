```python
%matplotlib notebook 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
xx, yy = np.meshgrid(x, y)

z = xx**4 - 6 * xx**2 * yy**2 + yy**4

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx,yy,z)
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-5,2)
plt.show()
```

```python
# fx_func = lambda x,y : 4*x**3 - 12*x*y**2
# fy_func = lambda x,y : 4*y**3 - 12*y*x**2

fx_func = lambda x,y : x * np.sign(x*y)
fy_func = lambda x,y : y * np.sign(x*y)
col_func = lambda x,y : -np.sign(x*y)

x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
xx, yy = np.meshgrid(x, y)
fx = fx_func(xx, yy)
fy = fy_func(xx, yy)
col = col_func(xx, yy)
norm = np.sqrt(fx**2 + fy**2)
fx /= norm
fy /= norm

plt.figure()
plt.quiver(xx, yy, fx, fy, col, scale=None)
plt.show()
```
