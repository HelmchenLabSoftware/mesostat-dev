```python
import numpy as np
import matplotlib.pyplot as plt

from mesostat.utils.plotting import custom_grad_cmap
```

```python
def plot_example(cmap):
    data = np.random.normal(0, 1, (50,50))
    plt.figure()
    plt.imshow(data, cmap=cmap)
    plt.colorbar()
    plt.show()

newcmp = custom_grad_cmap(
    np.array([
        [0,0,0],
        [255,198,126],
        [233,79,25]
    ]
))
    
plot_example(newcmp)
```
